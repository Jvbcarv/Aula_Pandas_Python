### **Manipulando colunas dentro da tabela**

```python
from google.colab import files
uploaded = files.upload()
```

```python
tabela = pd.read_csv("colunas_pandas.csv").head(5)
tabela
```

```python
#criando colunas dentro da tabela: Coloca o nome que quer dar para a coluna e atribui o valor que ela vai receber
#No exemplo abaixo coloquei "None", que significa "Nada" ou "Vazio"

tabela ["Coluna_nova"] = None
tabela
```

```python
tabela["Coluna_nova"] = "o que vc quiser colocar"
tabela
```

```python
#criando uma coluna e manipulando dados de uma coluna j√° existente para preencher a nova coluna:
#quero criar uma coluna de Sal√°rio Mensal na tabela

tabela ["Sal√°rio mensal"] = tabela ['Sal√°rio_Anual'] / 12
tabela
```

```python
#removendo uma coluna:
tabela = tabela.drop(columns=['Sal√°rio_Anual'])
tabela
```

```python
tabela = tabela.drop(columns=["ID_Departamento", 'Coluna_nova']) #d√° para apagar mais de uma coluna
tabela
```

```python
#Reorganizar colunas

tabela = tabela[["Ativo", "Nome", "Idade", "Sal√°rio", "Data_Registro"]]   #Basta escrever na ordem que voc√™ quer
tabela
```

```python
#Antes de reorganizar, √© √∫til visualizar as colunas (√∫til se a tabela for enorme)
#Como saber quais colunas existem?

print(tabela.columns)
```

```python
#Renomeando colunas:

#quero trocar a palavra "Nome" por "Funcion√°rio":

tabela.rename(columns = {"Nome": "Funcion√°rio"}, inplace=True)   #inplace=True ‚Üí altera direto no DataFrame (se remover, vai precisar salvar de novo em uma vari√°vel)
tabela
```

```python
#Renomear v√°rias colunas ao mesmo tempo

tabela.rename(columns={
    "Registro": "Admiss√£o",
    "ID_Departamento": "Departamento",
    "Ativo": "Status"}, inplace=True)
tabela
```

```python
#Renomear dados que est√£o dentro das colunas:
#quero trocar o "Sim" por Ativo, e o "N√£o" por Inativo.

tabela["Status"] = tabela["Status"].replace({"Sim": "Ativo", "N√£o": "Inativo"})
tabela
```

```python
df['Recebido'].std()  #std = Desvio Padr√£o (Standard Deviation)

# √â uma medida que mostra o quanto os valores est√£o espalhados em rela√ß√£o √† m√©dia.
# Desvio padr√£o baixo ‚Üí os valores est√£o pr√≥ximos da m√©dia, pouca varia√ß√£o.
# Desvio padr√£o alto ‚Üí os valores est√£o longe da m√©dia, muita varia√ß√£o.
```

```python
df['Recebido'].quantile([0.25, 0.50, 0.75])  #Quartis

#Imagine que voc√™ tem v√°rios valores e coloca todos em ordem (do menor para o maior).
#Os quartis dividem essa lista em 4 partes iguais:
# |----25%----|----25%----|----25%----|----25%----|
#       Q1          Q2          Q3

# 25%:	1¬∫ quartil do c√≥digo (Q1)	25% dos valores est√£o abaixo dele
# 50%:	2¬∫ quartil do c√≥digo (Q2) = mediana	Metade dos valores est√£o abaixo
# 75%:	3¬∫ quartil do c√≥digo (Q3)	75% dos valores est√£o abaixo
```

Interpretando o resultado dos quartis acima:

*   Q1 (0.25): 2.0 - 25% das vendas receberam 2 reais ou menos
*   Q2 (0.50): 10.0 -  50% das vendas receberam 10 reais  ou menos (este √© o valor mediano)
*    Q3 (0.75):  12.0 -  75% das vendas receberam R$ 12 reais  ou menos


```python
df.describe()   # O describe() j√° tr√°s m√©dia, std, min, max, quartis‚Ä¶

#Traz um resumo estat√≠stico das colunas num√©ricas: contagem (count), m√©dia (mean), desvio padr√£o (std), m√≠nimo (min), Q1, Q2, Q3 (quartis), m√°ximo (max)
#√â super usado para entender a distribui√ß√£o dos dados.
```

### **Contagem de Ocorr√™ncias**

```python
df = pd.read_excel("Papelaria_Vendas.xlsx")
df
```

```python
df['Item'].value_counts()
#contagem de quantas vezes cada valor aparece em uma coluna
#Muito √∫til para descobrir: Item mais vendido, Cidade com mais compras, Quais valores s√£o mais frequentes...
```

```python
df['Item'].value_counts().idxmax()

#idxmax(): Se aplicado em uma contagem, ele retorna o nome do item com o maior valor.
```

```python
df['Item'].value_counts().idxmin()

#idxmin(): Se aplicado em uma contagem, ele retorna o nome do item com o menor valor.
```

```python
print("Item mais vendido:", df['Item'].value_counts().idxmax())
print("Item menos vendido:", df['Item'].value_counts().idxmin())
print("Lugar que mais comprou:", df['Local da compra'].value_counts().idxmax())
print("Lugar que menos comprou:", df['Local da compra'].value_counts().idxmin())
```

Note que na resposta apareceu um "N√£o informado". Isso significa na coluna "Local da compra" tem campos com esse "N√£o informado".

```python
#Caso queira remover linhas com "N√£o informado":

df = df[df['Local da compra'] != 'N√£o informado']
```

```python
print("Lugar que mais comprou:", df['Local da compra'].value_counts().idxmax())
```

###**Opera√ß√µes de agrega√ß√£o**

**groupby()** √â uma forma de agrupar dados por uma coluna e aplicar opera√ß√µes de agrega√ß√£o em outra(s).

```python
df = pd.read_excel("Papelaria_Vendas.xlsx")
df
```

```python
#Quero ver o total de dinheiro recebido por cidade da tabela:

df.groupby('Local da compra')['Recebido'].sum()  #sum()=soma
```

```python
df.groupby('Item')['Recebido'].mean() #M√©dia recebida por item
```

```python
df.groupby('Item')['Recebido'].mean().astype(int) #coloquei o resultado em n√∫meros inteiros para visualizar melhor
```

###**Ordenando valores**

**.sort_values()** √© um m√©todo do Pandas para ordenar os valores

```python
#Quero ordenar os valores da coluna Recebido do maior para o menor

df.sort_values(by='Recebido', ascending=False).head(10)

#ascending=False: ordena do maior para o menor
# coloquei .head(10) no c√≥digo para ver os 10 primeiros
```

```python
df.sort_values(by='Recebido', ascending=True).head(10)
#ascending=False: ordena do menor para o maior
# coloquei .head(10) no c√≥digo para ver os 10 primeiros
```
